```python
import matplotlib.pyplot as plt
import numpy as np
from entangling_dropout.pipelines.data_processing.nodes import generate_data, NoiseType
```

```python
def generate_plot(x_values, y_values, x_scale, y_scale):
    # 100 linearly spaced numbers
    x = np.linspace(-1,1,100)

    # the function, which is y = sin(x) here
    y = np.sin(x*np.pi)

    # setting the axes at the centre
    fig = plt.figure()
    ax = fig.add_subplot(1, 1, 1)
    ax.spines['bottom'].set_position('center')
    ax.spines['left'].set_position('center')
    ax.spines['right'].set_color('none')
    ax.spines['top'].set_color('none')
    ax.xaxis.set_ticks_position('bottom')
    ax.yaxis.set_ticks_position('left')

    # plot the function
    plt.plot(x, y, 'b--')
    plt.plot(x_values*x_scale, y_values*y_scale, 'b.')

    # show the plot
    plt.show()
```

# Generating Amplitude Errors

```python
x_values, y_values, x_scale, y_scale = generate_data(samples=15, noise=0.3, seed=1337, noise_type=NoiseType.AMPLITUDE_ERROR)
```

```python
generate_plot(x_values, y_values, x_scale, y_scale)
```

```python
x_values, y_values, x_scale, y_scale = generate_data(samples=15, noise=0.3, seed=1337, noise_type=NoiseType.TIME_ERROR)
```
